### 1. Plot bar chart of CurrentJobTitleSelect

```python
plt.figure(figsize = (6,6))
df_mcq["CurrentJobTitleSelect"].value_counts().plot(kind="bar")
plt.show()
```

### 2. Order by frequency - descending

```python
df_mcq["CurrentJobTitleSelect"].value_counts().sort_values(ascending=False)
```

### 3. Order by frequency - ascending

```python
df_mcq["CurrentJobTitleSelect"].value_counts().sort_values(ascending=True)
```

### 4. Replace counts with percentages

```python
df_mcq["CurrentJobTitleSelect"].value_counts(normalize=True)*100 
```

### 5. Analyze the mean age by the Job titles.

```python
df_mcq.groupby("CurrentJobTitleSelect").agg({'Age': ['mean']})
```

### 6. Filter for rows where Age and CurrentJobTitleSelect are both not NA.

```python
df_mcq[(df_mcq["Age"].isna() == False) & (df_mcq["CurrentJobTitleSelect"].isna() == False)]
```

### 7. Group the dataset by CurrentJobTitleSelect to create a new variable mean_age, that is the mean age for each Job Title.

```python
df_mcq["mean_age"] = df_mcq.groupby("CurrentJobTitleSelect").transform('mean')
```

### 8. order of the plot to be "Entirely internal", "More internal than external","Approximately half internal and half external", "More external than internal", "Entirely external", "Do not know". Write your observation on nan values

```python
sns.countplot(x="WorkInternalVsExternalTools", data = df_mcq.fillna("NaN"),
              order = ["NaN","Entirely internal", "More internal than external","Approximately half internal and half external", "More external than internal", "Entirely external", "Do not know"])
plt.xticks(rotation=90)
plt.show()
```
