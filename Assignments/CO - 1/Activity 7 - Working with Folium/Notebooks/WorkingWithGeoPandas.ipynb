2. Read the data from the url :https://raw.githubusercontent.com/giswqs/data/main/world/world_cities.csv
Using markers show the population of the cities

```python
countries_world = pd.read_csv("https://raw.githubusercontent.com/giswqs/data/main/world/world_cities.csv")
countries_world.head()
```

```python
kol = folium.Map(location=[countries_world['latitude'][0], countries_world['longitude'][0]], tiles="openstreetmap", zoom_start=2)
for i in range(countries_world.shape[0]):
  folium.Marker(location=[countries_world['latitude'][i], countries_world['longitude'][i]],
                tooltip=countries_world['name'][i]+"  "+str(countries_world['population'][i])).add_to(kol)
kol
```

3. Run Dash to visualize the map at port 8050.

```python
import dash_html_components as html
import dash_leaflet as dl
from dash import Dash

# A few cities in Denmark
States = [dict(title=r["State"],position = [r["Latitude"],r["Longitude"]]) for i,r in state_wise.iterrows()]
# Create example app.
app = Dash()
app.layout = html.Div([
    dl.Map(children=[dl.TileLayer()] + [dl.Marker(**city) for city in States],
           style={'width': '100%', 'height': '50vh', 'margin': "auto", "display": "block"}, id="map"),
])

app.run_server()
```

5. 911 is an emergency telephone number of United States. Given a subset of 911 Emergency calls in a
flat file format “911.csv”, the task of a data scientist is defined below.

Using appropriate markers locate the five services reported by the public agencies of New York
in the year 2013 to visualize on open street map whose geo coordinate is
[40.2978759, -75.5812935]

```python
df_911 = pd.read_csv("/content/drive/MyDrive/Data Sciene Certification/DAY - 2/911.csv")
df_911.head()
```

```python
kol = folium.Map(location = [40.2978759, -75.5812935], tiles="openstreetmap",zoom_start=8)
for i in range(5):
  folium.Marker(location=[df_911['lat'][i], df_911['lng'][i]], tooltip=df_911['twp'][i]).add_to(kol)
kol
```
