### Q7. Consider the hypothesis " The sample is drawn from Gaussian Distribution". Report the test statistics on the hypothesis using the following normality test methods.

#### Shapiro-Wilk test

```python
# Shapiro-Wilk Test
# seed the random number generator
seed(1)
# generate univariate observations
data = norm_nums
# normality test
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

#### D'Agostino's K squared Test

```python
# D'Agostino and Pearson's Test
# seed the random number generator
seed(1)
# generate univariate observations
data = norm_nums
# normality test
stat, p = normaltest(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

### Anderson-Darling Test

### Q8.Generate three random Gaussian distribution as follows and plot  box and whiskers with five number summary.

```python
# Function for boxplot
def box_plot(mean,std,count):
    np.random.seed(10)

    # Generating data
    data = np.random.normal(mean, std, count)

    fig = plt.figure(figsize =(10, 7))

    # Creating plot
    plt.boxplot(data)

    # show plot
    plt.show()
    
def five_point_summary(data):
    five_point_summary = np.quantile(data,[0,0.25,0.5,0.75,1])
    print("Minimum :",five_point_summary[0])
    print("25 % Percentile",five_point_summary[1])   
    print("50 % Percentile",five_point_summary[2])
    print("75 % Percentile",five_point_summary[3])   
    print("Maximum :",five_point_summary[4])
```

```python
# 1. 1000 random numbers
box_plot(100,20,1000)
data = np.random.normal(100,20,1000)
five_point_summary(data)
```

```python
# 2. mean 0 and standard deviation 5.
box_plot(0,5,1000)
data = np.random.normal(0,5,1000)
five_point_summary(data)
```

```python
# 3. mean 0 and standard deviation 10.
box_plot(0,10,1000)
data = np.random.normal(0,10,1000)
five_point_summary(data)
```
