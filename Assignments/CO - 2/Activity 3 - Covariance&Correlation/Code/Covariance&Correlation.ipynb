### 1. Let page speed and purchase amount are totally random and independent of each other. What is your observation?

```python
%matplotlib inline

import numpy as np
from pylab import *
np.random.seed(8)
def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n-1)

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

scatter(pageSpeeds, purchaseAmount)

print("Covariance",covariance (pageSpeeds, purchaseAmount))

print("Correlation",np.corrcoef(pageSpeeds, purchaseAmount)[0][1])
```

### Covariance measures the direction of the relationship between two variables. Here negative covariance means page and purchase are negatively correlated

### Q2 What can be done to see the correlation between page speed and purchase amount (negative or positive correlation)?

```python
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y) / stddevx / stddevy 

correlation(pageSpeeds, purchaseAmount)
```

### It is a negative correlation

### 3. Consider income data centered around 25,000 with a normal distribution and standard deviation of 12,000, with 10,000 data points. Compute the mean.

```python
mu, sigma = 25000, 12000
s = np.random.normal(mu, sigma, 10000)
print("Mean",s.mean())
```

### 4. Segment the income data into 50 bins and plot the histogram.
