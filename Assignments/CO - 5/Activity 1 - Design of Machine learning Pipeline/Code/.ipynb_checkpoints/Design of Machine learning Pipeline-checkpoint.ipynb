```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import dython 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
df = pd.read_csv('../Data/employee_data.csv')
print(df.shape)
df.head()
```

```python
df['recently_promoted'].fillna(value=0, inplace=True)
```

```python
df["filed_complaint"].fillna(value=0, inplace=True)
```

```python
df.shape
```

Inspect the distribution of numeric and categorical feature individually and find the correlation between numerical features

```python
# encode nominal data
df['department'] = df['department'].astype('category')
df['department'] = df['department'].cat.codes
df['salary'] = df['salary'].astype('category')
df['salary'] = df['salary'].cat.codes
df['status'] = df['status'].astype('category')
df['status'] = df['status'].cat.codes
```

```python
df.describe()
```

```python
for i in df.columns:
    sns.displot(df[i],kde=True)
    plt.show()
```

```python
# drop categorical data
df.drop(['department','salary','status'],axis=1,)
# correlation between variables
corr = df.drop(['department','salary','status'],axis=1,).corr()
sns.heatmap(corr, annot=True)
```

```python


df['last_evaluation_missing'] = df['last_evaluation'].isnull()
df['last_evaluation_missing'] = df['last_evaluation_missing'].astype('int')
df['last_evaluation_missing'] = df['last_evaluation_missing'].astype('category')
df['last_evaluation_missing'] = df['last_evaluation_missing'].cat.codes

df['satisfaction_level_missing'] = df['satisfaction'].isnull()
df['satisfaction_level_missing'] = df['satisfaction_level_missing'].astype('int')
df['satisfaction_level_missing'] = df['satisfaction_level_missing'].astype('category')
df['satisfaction_level_missing'] = df['satisfaction_level_missing'].cat.codes

```

2. Create an indicator feature as per the condition stated below.
    1. 'underperformer' - last_evaluation < 0.6 and last_evaluation_missing == 0
    2. 'unhappy' - satisfaction_level < 0.2
    3. 'overachiever' - last_evaluation > 0.8 and satisfaction > 0.7
    
Finally, save the dataframe to analytical_table.csv

```python
#Create an indicator feature as per the condition stated below.
# 1. 'underperformer' - last_evaluation < 0.6 and last_evaluation_missing == 0
# 2. 'unhappy' - satisfaction_level < 0.2
# 3. 'overachiever' - last_evaluation > 0.8 and satisfaction > 0.7
# Finally, save the dataframe to analytical_table.csv
df['underperformer'] = np.where(df['last_evaluation'] < 0.6, 1, 0)
df['unhappy'] = np.where(df['satisfaction'] < 0.2, 1, 0)
df['overachiever'] = np.where(df['last_evaluation'] > 0.8, 1, 0)
df['overachiever'] = np.where(df['satisfaction'] > 0.7, 1, 0)
df['overachiever'] = np.where(df['last_evaluation_missing'] == 1, 0, df['overachiever'])
df['underperformer'] = np.where(df['satisfaction_level_missing'] == 1, 0, df['underperformer'])
df['unhappy'] = np.where(df['satisfaction_level_missing'] == 1, 0, df['unhappy'])

```

```python
df.head()
```

```python
# Merge underperformer, unhappy, overachiever with the dataframe into  single column 
df['indicator'] = np.where(df['underperformer'] == 1, 'underperformer', np.where(df['unhappy'] == 1, 'unhappy', np.where(df['overachiever'] == 1, 'overachiever', 'other')))
```

```python
df["indicator"] = df["indicator"].astype('category')
df["indicator"] = df["indicator"].cat.codes
df.head()
```

```python
df.to_csv('analytical_table.csv')
```

3. Using sklearn setup machine learning pipelines with following models,
    1. Logistic regression
    2. Random Forest classifier
    3. Gradient Boosting classifier.
    
Define hyperparameters to fit and tune each model using cross validation

```python
f_train_test = df.drop(['overachiever','underperformer','unhappy'],axis=1).dropna()
X_train, X_test, y_train, y_test = train_test_split(f_train_test.drop("indicator",axis=1), f_train_test['indicator'], test_size=0.2, random_state=42)
```

```python
# Logistic model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Random Forest model
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
print('Accuracy of random forest classifier on test set: {:.2f}'.format(random_forest.score(X_test, y_test)))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Gradient Boosting model
gradient_boost = GradientBoostingClassifier(n_estimators=100)
gradient_boost.fit(X_train, y_train)
y_pred = gradient_boost.predict(X_test)
print('Accuracy of gradient boosting classifier on test set: {:.2f}'.format(gradient_boost.score(X_test, y_test)))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
