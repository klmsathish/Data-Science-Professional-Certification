### Q7. Consider the hypothesis " The sample is drawn from Gaussian Distribution". Report the test statistics on the hypothesis using the following normality test methods.

#### Shapiro-Wilk test

```python
# Shapiro-Wilk Test
# seed the random number generator
seed(1)
# generate univariate observations
data = norm_nums
# normality test
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

#### D'Agostino's K squared Test

```python
# D'Agostino and Pearson's Test
# seed the random number generator
seed(1)
# generate univariate observations
data = norm_nums
# normality test
stat, p = normaltest(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```
